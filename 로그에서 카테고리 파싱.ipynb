```python
raw_url_list = df['Click Path(Premium)']
url_list=[x.split('>')for x in raw_url_list]

for i in range(len(url_list)):
    temp=np.array([])
    for j in range(len(url_list[i])):
        url_list[i][j] = url_list[i][j].strip()
        url_list[i][j]=url_list[i][j][25:]
        url_list[i][j] = url_list[i][j].strip()
        # home 카테고리 추가
        if url_list[i][j][0] =='/':
            url_list[i][j]=url_list[i][j].replace('/','')
        if url_list[i][j][0]=='(':
            url_list[i][j] = 'home '+url_list[i][j]
#         ?부터 (seconds)까지 제거 + ?~( 까지 문자열을 임시 문자열에 담고 그 후에 다시 삽입
        a = url_list[i][j].find('?')
        b = url_list[i][j].find('(')
        if a != -1 and b != -1 :
            url_list[i][j] = url_list[i][j][:a] + url_list[i][j][b-1:]
        # 임시배열 생성 후 순서대로 담고 이후에 결합
        temp = url_list[i][j].split()
        temp[1] = temp[1].replace('(','')
        temp = temp[:2]
        url_list[i][j] = temp
url_list
```
