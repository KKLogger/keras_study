```python
# DATA SET LOAD
from tensorflow.keras.datasets.boston_housing import load_data

(x_train,y_train),(x_test,y_test) = load_data(path="boston_housing.npz",
                                             test_split=0.2,
                                             seed=777)
print(x_train.shape,x_test.shape) # 13 개의 특성 / 각 특성의 값의 범위가 서로 다르다
print(y_train.shape,y_test.shape)
print(x_train[0])
print(y_train[0])
# 각 특성의 스케일이 다르다 -> 신경망의 성능에 영향을 준다 -> 범위가 다를 땐 범위가 동일하도록 조정이 필요 -> 표준화 작업
```

```python
# 표준화 ( Standardization) : 각 데이터의 특성에 대해 특성의 평균을 빼고 표준편차로 나눈다 -> 특성의 평균 = 0 특성의 표준편차 = 1 

import numpy as np

mean = np.mean(x_train,axis=0)
std = np.std(x_train,axis=0)

x_train = (x_train-mean)/std
x_test = (x_test-mean)/std
```

```python
# K-fold 
from sklearn.model_selection import KFold

k=3
#데이터셋을 k만큼 등분
#k-1 만큼 학습데이터 셋으로 1만큼 검증셋으로 이용
kfold = KFold(n_splits=k,random_state=777,shuffle=True)
```

```python
#CREATE MODEL
def get_model():
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense
    model = Sequential()
    model.add(Dense(64,activation="relu",input_shape=(13,)))
    model.add(Dense(32,activation="relu"))
    model.add(Dense(1))
    
    model.compile(optimizer="adam",loss="mse",metrics=['mae'])
    
    return model
mae_list=[] # 테스트셋 평가 후 결과 mae를 담는 리스트

for train_index, val_index in kfold.split(x_train): # k 번 진행
    #인덱스는 무작위로 추출됨
    #무작위로 생성되어 Overfitting을 방지
    x_train_fold , x_val_fold = x_train[train_index],x_train[val_index]
    y_train_fold , y_val_fold = y_train[train_index],y_train[val_index]
    
    model=get_model() #model 을 생성하는 함수가 3번 호출되며 3개의 모델이 생성된다.
    
    model.fit(x_train_fold,y_train_fold,
             epochs=300,
             validation_data = (x_val_fold,y_val_fold))
    _, test_mae = model.evaluate(x_test,y_test)
    mae_list.append(test_mae)
```

```python
print(mae_list)
```

```python
print(np.mean(mae_list)) #교차검증 (K-fold) 를 통해 모델의 성능이 향상되었다. 이보다 더 중요한건 데이터의 특성을 잘 파악하고 모델링 하는 것 이다,.
```
