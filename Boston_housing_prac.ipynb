```python
# DATA SET LOAD
from tensorflow.keras.datasets.boston_housing import load_data

(x_train,y_train),(x_test,y_test) = load_data(path="boston_housing.npz",
                                             test_split=0.2,
                                             seed=777)
print(x_train.shape,x_test.shape) # 13 개의 특성 / 각 특성의 값의 범위가 서로 다르다
print(y_train.shape,y_test.shape)
print(x_train[0])
print(y_train[0])
# 각 특성의 스케일이 다르다 -> 신경망의 성능에 영향을 준다 -> 범위가 다를 땐 범위가 동일하도록 조정이 필요 -> 표준화 작업
```

```python
# 표준화 ( Standardization) : 각 데이터의 특성에 대해 특성의 평균을 빼고 표준편차로 나눈다 -> 특성의 평균 = 0 특성의 표준편차 = 1 

import numpy as np

mean = np.mean(x_train,axis=0)
std = np.std(x_train,axis=0)

x_train = (x_train-mean)/std
x_test = (x_test-mean)/std
```

```python
# Create Validation Data

from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val = train_test_split(x_train,y_train,
                                              test_size=0.33,
                                              random_state=777)
```

```python
#데이터가 복잡하지 않고 데이터의 수가 적은데 모델을 깊게 쌓으면 과대적합 발생률 증가

#CREATE MODEL
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(64,activation="relu",input_shape=(13,)))
model.add(Dense(32,activation="relu"))
model.add(Dense(1)) # activation Function 을 설정하지 않으면 자동으로 Linear 가 된다.

model.compile(optimizer="adam",loss="mse",metrics=['mae'])
```

```python
#학습과정
history = model.fit(x_train,y_train,
         epochs=300,
          validation_data=(x_val,y_val)
         )
```

```python
model.evaluate(x_test,y_test) # 예측 값과 실제 값은 +mae 에서 -mae 만큼 차이가 난다.
```
